```python
import requests
from bs4 import BeautifulSoup
```

```python
homeurl= "https://peraturan.bpk.go.id"
```

```python
r = requests.get(url="https://peraturan.bpk.go.id/Details/69790/perda-kota-jambi-no-2-tahun-2014",headers={'User-Agent': 'Mozilla/5.0'})
```

```python
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser') 
print(soup.prettify()) 
```

```python
#Cari link dengan kategori peraturan untuk download
for a in soup.find_all('a',href=True, attrs={"data-kategori":"Peraturan"}):
    x = a['href']
print(x)
```

```python
#Gabung link
link = homeurl+x
print(link)
```

```python
#Request PDF
pdf = requests.get(link, allow_redirects=True, headers={'User-Agent': 'Mozilla/5.0'})
print(pdf)
```

```python
#Download
open("test.pdf","wb").write(pdf.content)
```
